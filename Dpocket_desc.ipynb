```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
df1 = pd.read_csv("dpout_explicitp_trans.csv", header = 0)
df2 = pd.read_csv("final_dpocket_soluble.csv", header = 0)
```

```python
df1 = df1.drop(columns=df1.columns[:9])
df2= df2.drop(columns=df2.columns[:9])
```

```python
df1 = df1.drop(columns=df1.columns[-20:])
df2 = df2.drop(columns=df2.columns[-20:])
```

```python
df1
```

```python
df1 = df1.apply(pd.to_numeric, errors='coerce')
df2 = df2.apply(pd.to_numeric, errors='coerce')
```

```python
df1
```

```python
# Initialize empty lists to store the statistics
df1_stats = []
df2_stats = []
```

```python
# Loop through the columns of each dataframe
for col in df1.columns:
  # Skip the column if it contains all 0s
  if (df1[col] == 0).all():
    continue

  # Calculate the descriptive statistics for the current column in df1
  df1_stat = df1[col].describe()

  # Add the statistics to the list
  df1_stats.append(df1_stat)
```

```python
for col in df2.columns:
  # Skip the column if it contains all 0s
  if (df2[col] == 0).all():
    continue

  # Calculate the descriptive statistics for the current column in df2
  df2_stat = df2[col].describe()

  # Add the statistics to the list
  df2_stats.append(df2_stat)
```

```python
# Concatenate the lists of statistics into DataFrames
df1_stats = pd.concat(df1_stats, axis=1).T
df2_stats = pd.concat(df2_stats, axis=1).T
```

```python
df2_stats
```

```python
# Save the statistics to CSV files
df1_stats.to_csv("df1_stats.csv", index=False)
df2_stats.to_csv("df2_stats.csv", index=False)
```
